```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import unicodedata
import pandas as pd
```

Função que encripta usando a cifra de cesar, recebe uma string e a chavr

```python
def encripta(string,chave):
    encriptada = ""
    for i in range(len(string)):
        char = string[i].upper()
        encriptada += chr((ord(char) + chave-65) % 26 + 65)
    return encriptada
```

```python
def limpaTexto(string):
  s = ''
  for i in range(len(string)):
      if(i != " "):
        char = string[i].upper()
        s += char 
  return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')
```

Recebe os parametros e executa a cifragem

```python
text = input('Coloque o texto:', )
s = int(input('Coloque a chave:', ))

textoLimpo = limpaTexto(text)

cifra = encripta(textoLimpo,s)

print ("Text  : " + text)
print ("mudança : " + str(s))
print ("cifra: " + cifra)
```
