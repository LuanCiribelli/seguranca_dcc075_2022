```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
```

```python
def encripta(string,chave):
    encriptada = ""
 
    for i in range(len(string)):
        char = string[i].upper()
 
    
        encriptada += chr((ord(char) + chave-65) % 26 + 65)
 
   
 
    return encriptada
```

```python
text = input('Coloque o texto:', )
s = 5
cifra = encripta(text,s)
print ("Text  : " + text)
print ("Shift : " + str(s))
print ("Cipher: " + cifra)
```

```python
freq_caracter = Counter(cifra)
freq_caracter_text = Counter(text)
print (f"A frequência da letra '{cifra}' na cifra é :\n { str(freq_caracter)}")
print (f"A frequência da letra '{text}' no texto é :\n { str(freq_caracter_text)}")
```

```python
linhas, calores = zip(*Counter(cifra).items())

indexes = np.arange(len(linhas))
width = 1

plt.bar(indexes, calores, width)
plt.xticks(indexes + width * 0.5, linhas)
plt.show()
```
